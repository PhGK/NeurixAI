```python
import pandas as pd
```

```python
# Read Omics Expression Protein Coding Genes Data
rna_data = pd.read_csv('../../data/raw/depmap_public_23q4_primary_files/OmicsExpressionProteinCodingGenesTPMLogp1.csv', index_col=0)
rna_data = rna_data.astype('double')
rna_data.columns = rna_data.columns.str.split(' ').str[0]
rna_data.to_csv('../../data/preprocessed/rna_data.csv')
print(f"Number of NaNs in dataframe: {rna_data.isna().sum().sum()}")
rna_data
```

```python
# Read somatic mutation data
s_mutation_data = pd.read_csv('../../data/raw/depmap_public_23q4_primary_files/OmicsSomaticMutationsMatrixHotspot.csv', index_col = 0)
s_mutation_data.columns = [col + '_mut' for col in s_mutation_data.columns]
s_mutation_data.to_csv('../../data/preprocessed/s_mutation_data.csv')
print(f"Number of NaNs in dataframe: {s_mutation_data.isna().sum().sum()}")
s_mutation_data
```

```python
# Merge s_mutation_data with rna_data on index as key
combined_s_mutation_rna_data = rna_data.merge(s_mutation_data, how='inner', left_index=True, right_index=True)

print(f"Number of NaNs in dataframe: {combined_s_mutation_rna_data.isna().sum().sum()}")

# Set NaNs to zero
combined_s_mutation_rna_data.to_csv('../../data/preprocessed/combined_s_mutation_rna_data.csv')
combined_s_mutation_rna_data
```
