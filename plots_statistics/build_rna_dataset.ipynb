```python
data = pd.read_csv('../data/OmicsExpressionProteinCodingGenesTPMLogp1.csv', index_col=0)
data
```

```python
rna_data = data.astype('double')
rna_data.columns = rna_data.columns.str.split(' ').str[0]
rna_data.to_csv('../use_data/rna_data.csv')
rna_data
```

```python
ordered = pd.DataFrame({}, index=rna_data.index)
```

```python
#add mutation and CNA data
mutation_data = pd.read_csv('../data/Hotspot_Mutations.csv', index_col = 0)
mutation_data.columns = [col + '_mut' for col in mutation_data.columns]

mutation_data= ordered.merge(mutation_data, how='left', left_index=True, right_index=True)
mutation_data = mutation_data.fillna(0)
mutation_data.to_csv('../use_data/mutation_data.csv')
mutation_data
```

```python
cna_data = pd.read_csv('../data/Copy_Number_Public_23Q2.csv', index_col = 0)
cna_data = cna_data.dropna(axis=1)
cna_data.columns = [col + '_cna' for col in cna_data.columns]
cna_data
```

```python
all_data = rna_data.merge(mutation_data, how = 'inner', left_index=True, right_index=True)
#all_data = all_data.merge(cna_data, how = 'inner', left_index=True, right_index=True)
```

```python
all_data.to_csv('../use_data/all_molecular_data.csv')
```
