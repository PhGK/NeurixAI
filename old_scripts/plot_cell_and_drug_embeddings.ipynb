```python
import torch as tc 
import pandas as pd
import numpy as np
from NN import Interaction_Model
from data import MyDataSet
```

```python
ds = MyDataSet(nsplits = 5)
```

```python
relevant_drugs = ds.cases['DRUG'].drop_duplicates().reset_index(drop=True)
relevant_drugs
```

```python
drug_embeddings = ds.drug_embeddings.loc[:,np.array(relevant_drugs)]
drug_embeddings_tensor = tc.tensor(drug_embeddings.to_numpy()).float()
drug_embeddings
```

```python
fingerprints = ds.drug_fingerprints.transpose().loc[:,np.array(relevant_drugs)]
fingerprints_tensor = tc.tensor(fingerprints.to_numpy()).float()
fingerprints
```

```python
full_drug_embeddings = tc.cat((drug_embeddings_tensor, fingerprints_tensor), axis=0).t()
full_drug_embeddings.shape
```

```python
def load_model(fold):
    model = Interaction_Model(ds).eval()
    model.load_state_dict(tc.load('../depmap/results/model_params/model_params_fold'+str(fold)+'.pt'))
    return model

    
models = load_model(1) #[load_model(fold).eval() for fold in [0,1,2,3,4]]
```

```python
latent_representation = models.nn1.forward(full_drug_embeddings).detach().numpy()
```

```python
pd.DataFrame(latent_representation, index = fingerprints.columns).to_csv('./results/latent_representations_from_models/drug_embeddings.csv')
```
